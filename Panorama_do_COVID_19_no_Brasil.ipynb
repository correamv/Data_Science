# Panorama do COVID-19 no Brasil

Declarada oficialmente em Março de 2020 pela OMS, a pandemia do COVID-19 afetou a população e economia globais de forma sem precedentes no mundo moderno.

A COVID-19 é uma doença altamente infecciosa transmitida pelo ar, principalmente por gotículas de saliva emitidas por indivíduos contaminados durante a fala, tosses ou espirros. Além disso a transmissão pode ser feita pelo contato com superfícies contaminadas.


Os sintomas podem variar de pessoa para pessoa, carga viral, fatores de risco como obesidade, hipertensão e etc (uma lista completa pode ser encontrada [aqui](https://www.gov.br/saude/pt-br/coronavirus/atendimento-tratamento-e-fatores-de-risco)).

Observou-se que cerca de 80% dos casos são assintomáticos ou leves, não necessitando de tratamento hospitalar, sendo tratamentos paliativos suficientes.

No entanto, 15% das pessoas terão infecções graves e precisarão de oxigênio. O restante das pessoas, que representam 5%, serão classificadas como infecções muito graves e precisarão de ventilação assistida, por meio de respiradores mecânicos em ambiente hospitalar.

Além disso, descobriu-se que principalmente em pacientes em que a doença se manifestou de forma moderada a grave, a COVID-19 pode deixar sequelas por tempo indefinido, como falta de ar, dificuldade na cognição e/ou memória, perda de paladar e olfato, fibrose nos pulmões e rins, entre outros.

Além do gravíssimo impacto humano, o mundo sofre com os impactos econômicos causados pela pandemia. Para conter a disseminação da doença, medidas protetivas como lockdowns, precisaram ser feitas. Com isso, diversas indústrias tiveram que pausar suas atividades, causando um impacto na cadeia de suprimentos global, que persiste até os dias de hoje.

Com o objetivo de mensurar a disseminação do covid-19 no Brasil, irei realizar uma análise com base nos dados públicos da doença.

## Obtenção dos Dados


Como fonte dos dados irei utilizar o dataset disponibilizado pela ***Our World in Data***, um projeto da Global Change Data Lab, uma organização sem fins lucrativos nativa da Inglaterra. Especialmente ao longo da pandemia, este time foi responsável por consolidar diversos dados ao redor do mundo, sendo crucial para diversos times de pesquisadores ao redor do globo.

Para o caso específico do Brasil, os dados crus, foram coletados através do ministério da saúde e da Universidade Federal de Viçosa.

Apesar do grande esforço para se adquirir dados de qualidade, a falta de testes no Brasil no início da pandemia, e a subnotificação de casos devido as diversas particularidades de cada região do país, tornam difíceis a consolidação das informações. Mas ainda assim, conseguiremos ter uma visão apurada de como a pandemia se manisfestou no país.


### Dicionário de Variáveis

* `iso_code` - 	ISO 3166-1 alpha-3 – códigos de país de três letras. Observe que regiões definidas pela OWID (por exemplo, continentes como ' Europa ') contêm prefixo 'OWID_'.
* `continent` - 	Continente referente a localização geográfica
* `location` - 	Localização Geográfica.
* `date` - 	Data de observação
* `total_cases` - 	Total de casos confirmados de COVID-19. Contagens pode incluir casos prováveis, quando relatado.
* `new_cases` - 	Novos casos confirmados de COVID-19. Contagens podem incluir casos prováveis, quando relatado. Em raros casos em que nossa fonte relata uma mudança diária negativa para uma correção de dados, definimos isso métrica para NA.
* `new_cases_smoothed` - 	Novos casos confirmados de COVID-19 (suavizado de 7 dias). Contagens pode incluir casos prováveis, quando relatado.
* `total_deaths` - 	Total de mortes atribuídas ao COVID-19. Contagens pode incluir mortes prováveis, onde relatado.
* `new_deaths` - 	Novas mortes atribuídas ao COVID-19. Contagens podem incluir mortes prováveis, onde relatado. Em raros casos em que nossa fonte relata uma mudança diária negativa para uma correção de dados, definimos isso métrica para NA.
* `new_deaths_smoothed` - 	Novas mortes atribuídas ao COVID-19 (suavizado de 7 dias). Contagens pode incluir mortes prováveis, onde relatado.
* `total_cases_per_million` - 	Total de casos confirmados de COVID-19 por 1.000.000 de pessoas. Contagens pode incluir casos prováveis, onde relatado.
* `new_cases_per_million` - 	Novos casos confirmados de COVID-19 por 1.000.000 de pessoas. Contagens pode incluir casos prováveis, onde relatado.
* `new_cases_smoothed_per_million` - 	Novos casos confirmados de COVID-19 (suavizado em 7 dias) por 1.000.000 de pessoas. Contagens pode incluir casos prováveis, onde relatado.
* `total_deaths_per_million` - 	Total de mortes atribuídas ao COVID-19 por 1.000.000 de pessoas. Contagens pode incluir mortes prováveis, onde relatado.
* `new_deaths_per_million` - 	Novas mortes atribuídas ao COVID-19 por 1.000.000 de pessoas. Contagens pode incluir mortes prováveis, onde relatado.
* `new_deaths_smoothed_per_million` - 	Novas mortes atribuídas ao COVID-19 (suavizado de 7 dias) por 1.000.000 de pessoas. Contagens pode incluir mortes prováveis, onde relatado.
* `reproduction_rate` - 	Estimativa em tempo real do a eficaz taxa de reprodução (R) do COVID-19. Veja https://github.com/crondonm/TrackingR/tree/main/Estimates-Database
* `icu_patients` - 	Número de pacientes com COVID-19 em Unidades de terapia intensiva (UTIs) em um determinado dia.
* `icu_patients_per_million` - 	Número de pacientes com COVID-19 em Unidades de terapia intensiva (UTIs) em um determinado dia por 1.000.000 pessoas.
* `hosp_patients` - 	Número de pacientes com COVID-19 no hospital em um determinado dia.
* `hosp_patients_per_million` - 	Número de pacientes com COVID-19 no hospital em um determinado dia por 1.000.000 pessoas.
* `weekly_icu_admissions` - 	Número de pacientes com COVID-19 recentemente admitido para unidades de tratamento intensivo (UTIs) em uma determinada semana (data do relatório e a anterior a 6 dias).
* `weekly_icu_admissions_per_million` - 	Número de pacientes com COVID-19 recentemente admitido para unidades de tratamento intensivo (UTIs) em uma determinada semana por 1.000.000 de pessoas (data do relatório e a anterior a 6 dias)
* `weekly_hosp_admissions` - 	Número de pacientes com COVID-19 recentemente admitido para hospitais em uma determinada semana (data do relatório e a anterior a 6 dias)
* `weekly_hosp_admissions_per_million` - 	Número de pacientes com COVID-19 recentemente admitido para hospitais em uma determinada semana por 1.000.000 de pessoas (data do relatório e a anterior a 6 dias)
* `total_tests` - 	Total de testes para COVID-19
* `new_tests` - 	Novos testes para COVID-19 (somente calculado para consecutivos dias)
* `total_tests_per_thousand` - 	Total de testes para COVID-19 por 1.000 pessoas
* `new_tests_per_thousand` - 	Novos testes para COVID-19 por 1.000 pessoas
* `new_tests_smoothed` - 	Novos testes para COVID-19 (suavizado de 7 dias). Para países que não reportam dados de testes diariamente, assumimos que os testes tenham mudado igualmente diariamente sobre quaisquer períodos em que nenhum dado foi relatado. Isso produz uma série completa de números diários, que é então média ao longo de uma janela contínua de 7 dias
* `new_tests_smoothed_per_thousand` - 	Novos testes para COVID-19 (suavizado de 7 dias) por 1.000 pessoas
* `positive_rate` - 	A partilha de testes COVID-19 que são positivos, dados como uma média móvel de 7 dias (este é a inverso dos testes_per_case)
* `tests_per_case` - 	Testes conduzido por novo caso confirmado de COVID-19, dado como uma média móvel de 7 dias (este é a inverso do positive_rate)
* `tests_units` - 	Unidades usadas por a localização para relatar seus dados de teste. Um arquivo de país não pode conter unidades misturadas. Todas as métricas relativas aos dados de teste usam as unidades especificadas. Unidades válidas são ' pessoas testadas ' (número de pessoas testadas), ' testes realizados ' (número de testes realizados. Uma única pessoa pode ser testada mais de uma vez em determinado dia) e 'amostras testadas ' (número de amostras testadas. Em alguns casos, mais de uma amostra pode ser requerido para realizar um determinado teste.)
* `total_vaccinations` - 	Número total das doses de vacinação COVID-19 administradas
* `people_vaccinated` - 	Número total de pessoas que tenham recebido ao menos 1 dose de vacina
* `people_fully_vaccinated` - 	Número total de pessoas que tenham recebido todas as doses prescritas pôr a inicial vacinação protocolo
* `total_boosters` - 	Número total de doses de reforço administradas (doses administradas além do número prescrito pelo protocolo de vacinação)
* `new_vaccinations` - 	Doses de vacinação de COVID-19 administradas (somente calculado para consecutivos dias)
* `new_vaccinations_smoothed` - 	Doses de vacinação COVID-19 administradas (suavizado de 7 dias). Para países que não relatório dados de vacinação por dia base, assumimos que _ vacinação mudado igualmente diariamente _ _ base sobre quaisquer períodos em que nenhum dado foi relatado. este produz uma série completa de números diários, que é então média ao longo de uma janela contínua de 7 dias
* `total_vaccinations_per_hundred` - 	Número total de doses de vacinação COVID-19 administradas por 100 pessoas na população total
* `people_vaccinated_per_hundred` - 	Número total de pessoas que tenham recebido ao menos 1 dose da vacina por 100 pessoas na população total
* `people_fully_vaccinated_per_hundred` - 	Número total de pessoas que tenham recebido todas as doses prescritas pôr a inicial vacinação protocolo por 100 pessoas na população total
* `total_boosters_per_hundred` - 	Número total de vacinação COVID-19 doses de reforço administradas por 100 pessoas na população total
* `new_vaccinations_smoothed_per_million` - 	Doses de vacinação COVID-19 administradas (suavizado em 7 dias) por 1.000.000 de pessoas na população total
* `new_people_vaccinated_smoothed` - 	Número diário de pessoas recebendo seus primeira dose da vacina (suavizado de 7 dias)
* `new_people_vaccinated_smoothed_per_hundred` - 	Número diário de pessoas recebendo seus primeira dose da vacina (suavizado de 7 dias) por 100 pessoas na população total
* `stringency_index` - 	Índice de Rigidez da Resposta do Governo: composto a medida Sediada em 9 indicadores de resposta incluindo fechamento de escolas, locais de trabalho e proibições de viagens, redimensionadas para um valor de 0 a 100 (100 = resposta mais rigorosa)
* `population` - 	População (últimos valores acessíveis). Consulte https://github.com/owid/covid-19-data/blob/master/scripts/input/un/population_latest.csv para obter a lista completa do fontes
* `population_density` - 	Número de pessoas dividido por area, medida em quilômetros quadrados, ano mais recente disponível
* `median_age` - 	Idade média da população, projeção da ONU para 2020
* `aged_65_older` - 	Parte da população que é 65 anos ou mais velha, ano mais recente disponível
* `aged_70_older` - 	Parte da população que é 70 anos ou mais velha, em 2015
* `gdp_per_capita` - 	Produto interno bruno na paridade de poder de compra (constante 2011 internacional dólares), ano mais recente disponível.
* `extreme_poverty` - 	Parte da população que vive em extrema pobreza, ano mais recente disponível, desde 2010
* `cardiovasc_death_rate` - 	Taxa de mortalidade por doença cardiovascular em 2017 (anual número de mortes por 100.000 pessoas)
* `diabetes_prevalence` - 	Prevalência de diabetes (% de população de 20 a 79 anos) em 2017
* `female_smokers` - 	Parte das mulheres que fumam, ano mais recente disponível
* `male_smokers` - 	Parte dos homens quem fumam, ano mais recente disponível
* `handwashing_facilities` - 	Parte da população saneamento básico para lavar as mãos, ano mais recente disponível
* `hospital_beds_per_thousand` - 	Leitos hospitalares por 1.000 pessoas, ano mais recente disponível desde 2010
* `life_expectancy` - 	Expectativa de vida no nascimento em 2019
* `human_development_index` - 	Um índice composto que mede a média da conquista em três básicas dimensões do desenvolvimento humano — um longo e saudável vida, conhecimento e um padrão de vida decente. Valores para 2019, importados de http://hdr.undp.org/en/indicators/137506
* `excess_mortality` - 	Diferença percentual entre o número relatado semanal ou mensal, de mortes em 2020-2021 e o número projetado de mortes para o mesmo período baseado em anos anteriores. Para mais informação acesse: https://github.com/owid/covid-19-data/tree/master/public/data/excess_mortality
* `excess_mortality_cumulative` - 	Diferença percentual entre o número relatado semanal ou mensal, de mortes desde 1 de janeiro de 2020 e o número cumulativo projetado de mortes para o mesmo período baseado em anos anteriores. Para mais informação acesse: https://github.com/owid/covid-19-data/tree/master/public/data/excess_mortality
* `excess_mortality_cumulative_absolute` - 	Diferença percentual cumulativa entre o número reportado de mortes desde 1 de janeiro de 2020 e o número projetado de mortes para o mesmo período baseado em anos anteriores. Para mais informação acesse: https://github.com/owid/covid-19-data/tree/master/public/data/excess_mortality
* `excess_mortality_cumulative_per_million` - 	Diferença cumulativa entre o número reportado de mortes desde o dia 1 de janeiro de 2020 e o número projetado de mortes para o mesmo período baseado em anos anteriores, por milhão de pessoas.Para mais informação acesse: https://github.com/owid/covid-19-data/tree/master/public/data/excess_mortality




### Importando os dados do COVID-19

* Usaremos este [link](https://github.com/owid/covid-19-data/tree/master/public/data) para importar a base de dados mais atualizada. Usaremos o formato `csv` neste projeto.

A data em que a base de dados foi extraída é de: 20/09/2022.

```
# importar as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt

# importar o dataset COVID-19 para um DataFrame
DATA_PATH = "https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv"
df = pd.read_csv(DATA_PATH)
```

Agora, vejamos para a data em que este projeto está sendo escrito, quais são os países com mais casos e com mais mortes devido a pandemia.


```
#5 países com mais casos de COVID-19 no mundo:

df.loc[df.date == "2022-09-20"].sort_values(by="total_cases", ascending= False)[0:10]
```

Buscando na documentação da base de dados e no dicionário de variáveis, vemos que na coluna `iso_code` as linhas que começam com o prefixo "OWID", são referentes a regiões maiores do que países, como por exemplo continentes, áreas de alta concentração de renda e etc.

Como neste momento só queremos olhar para os países especificamente, irei remover as linhas que contém este prefixo.

```
#removendo linhas que contém o prefixo "OWID" na coluna iso_code

#criando nova coluna auxiliar para identificar o prefixo OWID
df['nao_pais'] = df['iso_code'].str.startswith('OWID')

#filtrando do dataframe somente as linhas que contem o valor Falso na coluna "nao_pais"
df_pais = df[df["nao_pais"] == False]
```

```
del df_pais['nao_pais']
```

Agora sim, podemos identificar quais os países com mais casos e mortes por COVID-19.

```
df_pais.loc[df.date == '2022-09-20'].sort_values(by='total_cases', ascending= False)[0:5]
```

Podemos ver com base nas informações acima, que os cinco países com mais casos, são respectivamente:

*   Estados Unidos: 95.776.309 casos
*   Índia: 44.545.011 casos
*   França: 35.035.598 casos
*   Brasil: 34.592.027 casos
*   Alemanha: 32.797.308 casos



Agora os países com mais mortes:

```
df_pais.loc[df.date == '2022-09-20'].sort_values(by='total_deaths', ascending= False)[0:5]
```

```
top5_total_deaths = df_pais.loc[df.date == '2022-09-20', ['location', 'total_deaths']].sort_values(by="total_deaths", ascending=False)[0:5]

fig, ax = plt.subplots(figsize=(15,10))

ax.bar('location', 'total_deaths',data = top5_total_deaths)

plt.tight_layout()
```

Veja que não são os mesmos países que encontramos com o maior número de casos. Um fator que contribuiu muito para isso foi o ínício rápido da vacinação,  principalmente em países desenvolvidos na Europa.


Agora vejamos como o volume dos casos de COVID-19 se disseminaram ao longo da pandemia no Mundo, até o dia de hoje:

```
# gráfico de linha para mortes no mundo

df.loc[df.location == "World", "total_deaths"].reset_index(drop=True).plot(figsize=(10,8));
```

Podemos ver através do gráfico de linhas, que o número de mortes tende a se estabilizar, porém ainda não acabou por completo.

```
df_pais['gdp_per_capita'].describe()
```

Um gráfico interessante que podemos plotar é o de dispersão, considerando o produto interno bruto per capita de cada país, versus o número de mortes:

```
fig2, ax2 = plt.subplots(figsize=(15,13))

df_pais[df_pais.date =='2022-09-20'].plot.scatter(x='gdp_per_capita', y='total_deaths', ax=ax2, c='Blue')
ax2.set_xscale('log')
ax2.set_yscale('log')
ax2.set_title("GDP per capita vs Número de Mortes - Escala Logarítmica")

plt.tight_layout()
```

Podemos observar que mesmo países com um PIB maior, tiveram um grande número de mortos pela COVID-19. O que indica que somente o desenvolvimento econômico do pais não é um fator decisivo em tempos de pandemia. A situação política, cultural, a presença ou não de um sistema público de saúde, entre outros fatores, precisam ser avaliados com igual importância.

### Análise Exploratória para o Brasil


Já que tivemos um panorama de como está o Mundo, vamos olhar especificamente para o Brasil.
Para isso vamos isolar os dados referentes ao Brasil em um dataframe específico.

```
#criando dataframe para o Brasil
df_brasil = df.loc[df.location == "Brazil"].copy()

df_brasil.head()
```

Vamos verificar agora quando ocorreram o primeiro caso e a primeira morte no Brasil.

```
#qual foi a data do primeiro caso no Brasil?
primeiro_caso = df_brasil.loc[df_brasil.new_cases != 0.0, ['date', 'new_cases']].sort_values(by="date", ascending=True).head(1)
primeiro_caso
print("O primeiro caso identificado no Brasil ocorreu no dia: {} ".format(df_brasil.date.loc[27104]))
```

```
#qual foi a data da primeira morte no Brasil?
primeira_morte = df_brasil.loc[(df_brasil.new_deaths != 0.0) & (df_brasil.new_deaths.isnull() == False)].head(1)
print("A primeira morte identificada ocorreu no dia {}".format(df_brasil.date.loc[27124]))
```

Quanto tempo decorreu entre a primeira morte e o primeiro caso?



```
# Identificar quantos dias demorou para registrarmos a primeira morte, considerando o primeiro caso de COVID-19 registrado.
# calculando tempo desde o primeiro caso até a primeira morte

from datetime import date

delta = df_brasil.date.loc[27124] - df_brasil.date.loc[27104]
print("Entre o primeiro caso e a primeira morte no Brasil se passaram {} dias.".format(delta.days))
```

Para finalizar a análise dos dados, irei plotar os gráficos de casos totais e mortes no Brasil, em escala linear e logarítmica.

```
# Plotar gráficos de linha em escalas linear e logarítmica para os casos totais e mortes totais.

x = list(range(939))
y = df_brasil['total_cases']
y2 = df_brasil['total_deaths']

fig, ((ax1,ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(15,12))

#casos totais em escala linear
ax1.set_title("Casos no Brasil")
ax1.set_xlabel("Nº Dias")
ax1.set_ylabel("Número de Casos")
ax1.plot(x,y)

#casos totais em escala logarítmica
ax2.set_title("Casos no Brasil - Escala Logarítmica")
ax2.set_xlabel("Nº Dias")
ax2.set_ylabel("Número de Casos")
ax2.set_yscale('log')
ax2.plot(x,y)

#mortes totais em escala linear
ax3.set_title("Mortes no Brasil")
ax3.set_xlabel("Nº Dias")
ax3.set_ylabel("Número de Mortes")
ax3.plot(x,y2)

#casos totais em escala logarítmica
ax4.set_title("Mortes no Brasil - Escala Logarítmica")
ax4.set_xlabel("Nº Dias")
ax4.set_ylabel("Número de Mortes")
ax4.set_yscale('log')
ax4.plot(x,y2)

plt.tight_layout()
```
